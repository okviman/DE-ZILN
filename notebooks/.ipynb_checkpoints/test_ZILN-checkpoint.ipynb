```python
import numpy as np
import pandas as pd
import scipy.stats as stats
```

```python
X = pd.read_csv("test/x.csv", header=None, index_col=0)
Y = pd.read_csv("test/y.csv", header=None, index_col=0)
X_np = X.to_numpy()
Y_np = Y.to_numpy()
gene_count = X_np.shape[0]
print(gene_count)
```

```python
# x numpy vector of counts
# transformation to apply to the non-zero elements of x.
def process_counts(x, transformation):
    cell_count = len(x)
    N_0 = np.sum(x == 0)
    N_plus = cell_count - N_0
    t_x = transformation(x[np.where(x > 0)])    
    return(t_x, N_plus, N_0)
```

```python
ci_x = list(map(lambda x: utils.get_intervals(*process_counts(x, np.log), z=1.96, model="lognormal"), X_np))
```

```python
ci_y = list(map(lambda x: utils.get_intervals(*process_counts(x, np.log), z=1.96, model="lognormal"), Y_np))
```

```python
# return p-value -- we are using Normal assumption on the means for inference. The null is that the means are the same.
def infer_lfc(m_x, s_x, m_y, s_y):
    diff = m_x - m_y 
    s_hat = np.sqrt((s_x**2) + (s_y**2))
    lb = mu_hat - 1.96*s_hat
    ub = mu_hat + 1.96*s_hat
    # Compute the p-value.
    z_score = diff / s_hat
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))
    return(lb,ub,p_value)
```

```python
infer_results = [infer_lfc(xx[1], xx[2], yy[1], yy[2]) for xx, yy in zip(ci_x, ci_y)]
infer_results = np.asarray(infer_results)
```
